```python
# Importing libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
# read in the data
try:
    df = pd.read_csv(r'C:\Users\user\Downloads\movies.csv\movies.csv')
    print("File Loaded")
except:
    print("Error")

```

```python
# First look at the data
df.head()
```

```python
# Looking for missing data

df.isna().sum()
```

```python
# Removing missing rows

df = df.dropna()
```

```python
# Data types
df.dtypes
```

```python
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
pd.set_option('display.max_rows',None)
```

```python
df.head().sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
# Dropping duplicates (If requiered)

# df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Checking the correlation between the budget and the gross revenue

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs gross (In millions)')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for film')
```

```python
# Plot budget vs gross using seaborn 

sns.regplot( x='budget',y='gross',data=df,scatter_kws={"color":'red'},line_kws={"color":"blue"} )
```

```python
# Showing positive correlation between the budget and the gross revenue

df.corr(method='pearson')
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True) # Black / Dark colors - low correlation , Bright colors - High correlation
plt.title('Correlation Matrix For Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Looking at the company column

df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True) # Black / Dark colors - low correlation , Bright colors - High correlation
plt.title('Correlation Matrix For Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
sorted_pairs = corr_pairs.sort_values()
sorted_pairs[(sorted_pairs)>0.5] # Showing pairs with a correlation higher than 0.5
```
